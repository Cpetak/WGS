```python
I want a column: env_var with ph_mean,ph_freq etc., a column: below or above, a column: snp
```

```python
ndf
```

```python
phm=ndf[["ph_mean_cat","snp_count"]].copy()
phm["env_var"]="ph_mean_cat"
phm=phm.rename(columns={"ph_mean_cat": "bayenv"})
```

```python
phf=ndf[["ph_freq_cat","snp_count"]].copy()
phf["env_var"]="ph_freq_cat"
phf=phf.rename(columns={"ph_freq_cat": "bayenv"})
```

```python
tm=ndf[["temp_mean_cat","snp_count"]].copy()
tm["env_var"]="temp_mean_cat"
tm=tm.rename(columns={"temp_mean_cat": "bayenv"})
```

```python
tv=ndf[["temp_var_cat","snp_count"]].copy()
tv["env_var"]="temp_var_cat"
tv=tv.rename(columns={"temp_var_cat": "bayenv"})
```

```python
tv
```

```python
result = pd.concat([phm,phf,tm,tv])
```

```python
result.reset_index()
```

```python
ax = sns.boxplot(x="bayenv", y="snp_count",data=result, palette="Set3")
```

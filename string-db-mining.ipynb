1) with request, however, we need to know the url our protein corresponds to.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://string-db.org/cgi/network?taskId=bCxbg9pRVvSS"
back = requests.get(url)
```

```python
soup=BeautifulSoup(back.text, "html.parser")
```

```python
line =soup.find(style="display:table-cell;text-align:left;padding-left:10px;")
```

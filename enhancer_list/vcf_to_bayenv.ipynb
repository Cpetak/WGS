```python
import pandas as pd
import numpy as np
```

```python
loc = pd.read_table("my_loc.txt", sep="\t", header=None)
```

```python
loc = loc.replace(np.nan, ' ', regex=True)
```

```python
loc
```

```python
def row_from_row(pops):
    refs=[]
    alts=[]
    for p in pops:
        t = p.to_numpy()
        te = [ i.split("/") for i in t]
        te = np.delete(te, np.where(te == " "))
        if len(te) < 40:
            te=np.concatenate(te, axis=0 )
            te = np.delete(te, np.where(te == " "))
        temp=[int(i) for i in te]
        ref=len(temp)-sum(temp)
        alt=sum(temp)
        refs.append(ref)
        alts.append(alt)
    return(alts,refs)
```

```python
for i in range(len(loc)):
    pop1=loc.iloc[i,9:29]
    pop2=loc.iloc[i,29:49]
    pop3=loc.iloc[i,49:69]
    pop4=loc.iloc[i,69:89]
    pop5=loc.iloc[i,89:109]
    pop6=loc.iloc[i,109:129]
    pop7=loc.iloc[i,129:149]
    pops=[pop1,pop2,pop3,pop4,pop5,pop6,pop7]
    alts,refs=row_from_row(pops)
    if i == 0:
        df = pd.DataFrame([alts])
        df = df.append([refs],ignore_index = True) 
    else:
        df = df.append([alts],ignore_index = True)
        df = df.append([refs],ignore_index = True) 
```
